# Fourier transform

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as nd
import cv2

img1 = np.array([[0, 255],
                [0, 255]])

print(img1)

plt.imshow(img1, cmap='gray')
plt.show()
```

```python
%matplotlib inline
fft_img1 = np.fft.fft2(img1)/4
fft_img1 = np.fft.fftshift(fft_img1)
print(fft_img1)
magnitude_spectrum = np.abs(fft_img1)

plt.imshow(magnitude_spectrum, cmap='gray')
plt.show()
```

```python
img2 = cv2.imread("./images/denoising/BSE_noisy.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(img2, cmap="gray")
plt.show()
```

#  Image smoothing : gaussian, median

```python
fft_img2 = np.fft.fft2(img2)
fft_img2 = np.fft.fftshift(fft_img2) #Shifting zero freq to center of image

magnitude_spectrum = 20*np.log(np.abs(fft_img2))

plt.imshow(magnitude_spectrum, cmap="gray")
plt.show()
```

## LPF

```python
rows, cols = img2.shape
crow, ccol = int(rows / 2), int(cols / 2) #center row, center col

mask = np.zeros((rows, cols), np.uint8)
r = 100
center = [crow, ccol]
x, y = np.ogrid[:rows, :cols]
mask_area = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= r*r #true false numpy array

mask[mask_area] = 1


#Apply mask and draw new magnitude spectrum
lpf_img2 = fft_img2*mask
lpt_mag_spectrum = 2000*np.log(np.abs(lpf_img2))

plt.imshow(lpt_mag_spectrum, cmap="gray")
plt.show()
```

```python
#Reverse fft
reverse_img2 = np.fft.ifftshift(lpf_img2)
reverse_img2 = np.fft.ifft2(reverse_img2)
reverse_img2 = np.abs(reverse_img2)

plt.imshow(reverse_img2, cmap="gray")
plt.show()
```

```python
def gaussian_kernel(size, size_y=None):
    size = int(size)
    if not size_y:
        size_y = size
    else:
        size_y = int(size_y)
    x, y = np.mgrid[-size:size+1, -size_y:size_y+1]
    g = np.exp(-(x**2/float(size)+y**2/float(size_y)))
    return g / g.sum()

gaussian_kernel_array = gaussian_kernel(1)
print(gaussian_kernel_array)
plt.imshow(gaussian_kernel_array, cmap=plt.get_cmap('jet'), interpolation='nearest')
plt.colorbar()
plt.show()
```
