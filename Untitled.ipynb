```python
import numpy as np
import matplotlib.pyplot as plt

img = np.array([[0, 255],
                [0, 255]])

print(img)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
%matplotlib inline
fft_img1 = np.fft.fft2(img)/4
fft_img1 = np.fft.fftshift(fft_img1)
print(fft_img1)
magnitude_spectrum = np.abs(fft_img1)

plt.imshow(magnitude_spectrum, cmap='gray')
plt.show()
```

#  Image smoothing : gaussian, medium, 

```python
def gaussian_kernel(size, size_y=None):
    size = int(size)
    if not size_y:
        size_y = size
    else:
        size_y = int(size_y)
    x, y = np.mgrid[-size:size+1, -size_y:size_y+1]
    g = np.exp(-(x**2/float(size)+y**2/float(size_y)))
    return g / g.sum()

gaussian_kernel_array = gaussian_kernel(1)
print(gaussian_kernel_array)
plt.imshow(gaussian_kernel_array, cmap=plt.get_cmap('jet'), interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
from scipy import ndimage as nd
from scipy import ndimage as nd

import cv2

img = cv2.imread("./images/denoising/noisy_img.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
gaussian_img = nd.gaussian_filter(img, sigma=3)

plt.imshow(gaussian_img, cmap='gray')
plt.show()
```

```python
median_img = nd.median_filter(img, size=3)
plt.imshow(median_img, cmap='gray')
plt.show()
```

```python
gaussian_img = nd.gaussian_filter(gaussian_img, sigma=3)

plt.imshow(gaussian_img, cmap='gray')
plt.show()
```
